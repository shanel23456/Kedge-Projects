## The first method using haversine  
### About the distance from each demand point in zone orgine to all the public transpotation stops 

```python
# example of calculating the distance from the first demand point to all tha public transpotation stops 
# choose the first demand point from data4(cleaned demand dataset)
import numpy as np
lat0 = np.full((2914,),data4["LAT_CENTRO_ETAB_Origin"].iloc[0])
long0 = np.full((2914,),data4["LNG_CENTRO_ETAB_Origin"].iloc[0])
```

```python
demandpoint0 = pd.DataFrame({'lat':lat0, 'lng':long0})
demandpoint0
```

```python
lat1 = np.array(public["LAT"])
lng1 = np.array(public["LNG"])
lat2 = np.array(demandpoint0["lat"])
lng2 = np.array(demandpoint0["lng"])
```

```python
EARTH_RADIUS = 6371

def haversine_distance(lat1, lng1, lat2, lng2):
    lat1, lng1, lat2, lng2 = map(np.radians, [lat1, lng1, lat2, lng2])
    
    lat = lat2 - lat1
    lng = lng2 - lng1
    d = np.sin(lat * 0.5) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(lng * 0.5) ** 2
    h_dist = 2 * EARTH_RADIUS * np.arcsin(np.sqrt(d))
    return h_dist
```

```python
public["distance0"] = haversine_distance(public["LAT"].values, public["LNG"].values, demandpoint0["lat"].values, demandpoint0["lng"].values)
```

```python
public
```

```python
# choose the distance <= 1.5 km 
dist0 = public.loc[public["distance0"]<=1.5]
dist0["distance0"].value_counts().sum()
# 34 possible stops near the first demand point
```

```python
d={}
dist={}
```

```python
for i in range(63,67): 
    
    lat = np.full((2914,),data4["LAT_CENTRO_ETAB_Origin"].iloc[i])
    long = np.full((2914,),data4["LNG_CENTRO_ETAB_Origin"].iloc[i])
    stop_name=public["stop_name"]
    occurence=public["occurence"]
    d[i]=pd.DataFrame({'stop':stop_name,'occurence': occurence, 'lat':lat, 'lng':long})
    
    d[i]["distance"] = haversine_distance(public["LAT"].values, public["LNG"].values, d[i]["lat"].values, d[i]["lng"].values)

    dist[i]= d[i].loc[d[i]["distance"]<=1.5]
    print(dist[i].value_counts().sum())
    print(dist[i])
```
