# Cleaning data

```python
import pandas as pd
solution_file = pd.read_excel("solution.xlsx")
pd.set_option("max_colwidth", None)
solution_file
```

```python
solution_file.info()
```

## Remove symbols and letters

```python
solution_file["departure"] = solution_file["departure"].str.replace("-", " ")
solution_file["departure"] = solution_file["departure"].str.replace("X", " ")
solution_file["departure"] = solution_file["departure"].str.replace("[", " ")
solution_file["departure"] = solution_file["departure"].str.replace("--", " ")
solution_file["departure"] = solution_file["departure"].str.replace(";", " ")
solution_file["departure"] = solution_file["departure"].str.replace("(", " ")
solution_file["departure"] = solution_file["departure"].str.replace("on", " ")
```

```python
solution_file["destination"] = solution_file["destination"].str.replace("TO", " ")
solution_file["destination"] = solution_file["destination"].str.replace(";", " ")
solution_file["destination"] = solution_file["destination"].str.replace("(", " ")
solution_file["destination"] = solution_file["destination"].str.replace("on", " ")
```

```python
solution_file["number of containers"] = solution_file["number of containers"].str.replace(",", " ")
solution_file["number of containers"] = solution_file["number of containers"].str.replace("of", " ")
solution_file["number of containers"] = solution_file["number of containers"].str.replace("container", " ")
solution_file["number of containers"] = solution_file["number of containers"].str.replace("cont", " ")
solution_file["number of containers"] = solution_file["number of containers"].str.replace(">", " ")
solution_file["number of containers"] = solution_file["number of containers"].str.replace("=", " ")
solution_file["number of containers"] = solution_file["number of containers"].str.replace("1", " ")
solution_file["number of containers"] = solution_file["number of containers"].str.replace("]", " ")
```

```python
solution_file
```

```python
solution_file.to_excel("New_solution.xlsx", index=False)
```
